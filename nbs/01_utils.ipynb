# utils

> Utility functions that can be used for plotting and stuff.

```python
#export
def plot_images(images):
    n = len(images)
    axs = get_grid(n, figsize=(12,9))
    for ax, im in zip(axs, images):
        if im.shape[0]==1:
            im = rearrange(im, 'c h w -> h w c')
            ax.imshow(im, cmap='gray')
        elif im.shape[0]==3:
            im = rearrange(im, 'c h w -> h w c')
            ax.imshow(im)
        else:
            ax.imshow(im)
        ax.axis('off')
    plt.show()
```

```python
images = [Image.open(path) for path in glob.glob('./data/samples/*.jpg')]
plot_images(images)
```
